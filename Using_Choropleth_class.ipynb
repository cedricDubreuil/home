```
import os
import folium
import json
import requests
import pandas as pd
import geopandas

from branca.colormap import linear
```

```
url = 'https://france-geojson.gregoiredavid.fr/repo/' # Carte de la France par départements
fr_state = f'{url}/departements.geojson'
map = folium.Map(location=[42, -0.5], zoom_start=5)
```

Méthode GeoJson string (geo_str)

```
folium.Choropleth(geo_data=fr_state).add_to(map)
```

Méthode filename (geo_path)

```
folium.Choropleth(
    geo_data=geopandas.read_file(fr_state),
    fill_opacity=0.3,
    line_weight=2,
).add_to(map)
```

Sauvegarde de la map

```
map.save(os.path.join(pathFile, 'map_name.html'))
```

```
df_election2017_2 = pd.read_csv(f'{pathFile}2017-2.csv', sep=';', encoding= 'cp1252')
df_election2017_2.head(3)
```

[ColorBrewer, Color Advice for Maps](https://colorbrewer2.org/#type=sequential&scheme=BuGn&n=3)

```
folium.Choropleth(
    geo_data=fr_state,
    data=df_election2017_2,
    key_on='feature.properties.code',
    columns=['Code', 'Abstentions'],
    fill_color= 'Blues',
    bins= 8,
    legend_name= 'Test', 
    overlay=True, 
    control=True,
).add_to(map)

# map.save(os.path.join(pathFile, 'nombre_inscrits.html'))
map
```
